```python
import findspark
findspark.init()

import pyspark
sc = pyspark.SparkContext.getOrCreate()

postsRDD = sc.textFile("PostHistory.xml").map(lambda x: x.split("\<row")[0]).map(lambda x: x.split(" ")).filter(lambda x: len(x) > 4).collect()
a = {}
for x in postsRDD: 
    if str(x[8]) not in a:
        a[str(x[8])] = 0
    counter = False
    for y in x:
        if y == "Text":
            counter = True
            a[str(x[8])]+=1
            continue
        elif y == "\/>":
            break
            
        if counter:
            a[str(x[8])]+=1

for key,value in a.items():
    print (key + " has: " + str(value) + " words")
```
